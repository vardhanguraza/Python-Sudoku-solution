```python
class sudoku:
    def __init__(self, sudoku_size, array = []):
        self.sudoku_size = sudoku_size
        self.sudoku_matrix = array
        self.valid_inputs = []
        self.missing_data_sector = {}
        self.missing_data_cell = {}
        self.sector_index_list = {} #Shows the list of all cells in a sector
        self.sector_update_flag = [] # Shows if any sector still needs to traverse
        self.print_values()
        
    def create_matrix(self):
        
        #Level 1 array insertion
        sudoku_level1_matrix = []
        for a in range(0, self.sudoku_size) :
            
            #Level 2 array insertion
            sudoku_level2_matrix = []
            for b in range(0, self.sudoku_size) :
                
                #Level 3 array insertion
                sudoku_level3_matrix = []
                for c in range(0, self.sudoku_size) :
                    
                    #Level 4 array insertion
                    sudoku_level4_matrix = []
                    for d in range(0, self.sudoku_size) :
                        sudoku_level4_matrix.append(input('Please enter number between 1 and %d : ' %(self.sudoku_size)))
                    
                    #Insert array data to level 3
                    sudoku_level3_matrix.append(sudoku_level4_matrix)
                    
                #Insert array data to level 2
                sudoku_level2_matrix.append(sudoku_level3_matrix)
                
            sudoku_level1_matrix.append(sudoku_level2_matrix)
        self.sudoku_matrix = sudoku_level1_matrix
            
            
        #Print the create matrix
        self.print_values()
            
    #Create an dictionary of missing data in sector
    def missing_data(self):
        sector_count = 0
        valid_inputs = []
        for a in range(0, self.sudoku_size):
            for b in range(0, self.sudoku_size):
                valid_inputs = [a for a in range(1, self.sudoku_size ** 2 + 1)]
                for c in range(0, self.sudoku_size):
                    for d in range(0, self.sudoku_size):
                        if self.sudoku_matrix[a][b][c][d] in valid_inputs:
                            valid_inputs.remove(self.sudoku_matrix[a][b][c][d])
                
                #Flag values to tell if the sector need to be updated
                if len(valid_inputs) > 0:  
                    self.sector_update_flag.append(1) #1 signifies needs update
                else :
                    self.sector_update_flag.append(0) #0 signifies needs update
                
                #Updating values of each cell in dictionary
                self.missing_data_sector.update({str(sector_count) : valid_inputs})
                self.sector_index_list.update({str(sector_count) : []})
                sector_count += 1
        sector_count = 0
        
        #Print the dictionary of missing data in an sector
        print('The missing data in each sector : ')
        for key in self.missing_data_sector:
            print(key + ' -> ', self.missing_data_sector[key])
        
        print('Which sectore needs update : ', )
        for a in range(0, len(self.sector_update_flag)):
            print("Sector -> ", a, "Update value = ", self.sector_update_flag[a])
        
        #Create missing data in each cell dictionary
        for a in range(0, self.sudoku_size):
            for b in range(0, self.sudoku_size):
                for c in range(0, self.sudoku_size):
                    for d in range(0, self.sudoku_size):
                        
                        self.sector_index_list[str(sector_count)].append(str(a)+str(b)+str(c)+str(d)) #Create an array of all cells in a sector
                        if self.sudoku_matrix[a][b][c][d] == 0 :
                            self.missing_data_cell.update({str(a)+str(b)+str(c)+str(d) : []})
                            #Using for loop to assign value to avoid same memory allocation while using the same array to assign values
                            for i in range(0, len(self.missing_data_sector[str(sector_count)])):
                                self.missing_data_cell[str(a)+str(b)+str(c)+str(d)].append(self.missing_data_sector[str(sector_count)][i])
                            
                        else :
                            self.missing_data_cell.update({str(a)+str(b)+str(c)+str(d) : [self.sudoku_matrix[a][b][c][d]]})
                sector_count += 1          
        print("Cells in each sector")
        for key in self.sector_index_list:
            print(key + " -> ", self.sector_index_list[key])
        print('The missing data in each cell')
        for key in self.missing_data_cell:
            print(key + " -> ", self.missing_data_cell[key])
        
    
    #
    def sector_values_check(self, sector_number):
        cell_indexes = self.sector_index_list[str(sector_number)]
        
        #First loop is to set the horizontal and vertical coordinates for each cell
        for x in range(0, len(cell_indexes)):
            h1 = int(cell_indexes[x][0])
            h2 = int(cell_indexes[x][2])
            v1 = int(cell_indexes[x][1])
            v2 = int(cell_indexes[x][3])
            cell_options = self.missing_data_cell[cell_indexes[x]]
            if len(cell_options) > 1 :
                remove_options = []
                for index in range(0, len(cell_options)):
                    found_value_flag = False
                    for a in range(0, self.sudoku_size):
                        for b in range(0, self.sudoku_size):

                            #Only check when there is only one value options in that cell
                            if len(self.missing_data_cell[str(a)+str(v1)+str(b)+str(v2)]) == 1:
                                #For vertical movement
                                #print(cell_options, index)
                                if self.missing_data_cell[str(a)+str(v1)+str(b)+str(v2)][0] == cell_options[index]:
                                    remove_options.append(cell_options[index])
                                    found_value_flag = True
                                    break
                                 
                            #Only check when there is only one value options in that cell
                            if len(self.missing_data_cell[str(h1)+str(a)+str(h2)+str(b)]) == 1:
                                #For horizontal movement
                                if self.missing_data_cell[str(h1)+str(a)+str(h2)+str(b)][0] == cell_options[index]:
                                    remove_options.append(cell_options[index])
                                    found_value_flag = True
                                    break
                        if found_value_flag:
                            break
                            
                if len(remove_options) > 0:
                    for a in range(0, len(remove_options)):
                        self.missing_data_cell[cell_indexes[x]].remove(remove_options[a])

                    if len(self.missing_data_cell[cell_indexes[x]]) == 1 :
                        for a in range(0, len(cell_indexes)):
                            if len(self.missing_data_cell[cell_indexes[x]]) > 1 and self.missing_data_cell[cell_indexes[x]][0] in self.missing_data_cell[cell_indexes[a]]:
                                self.missing_data_cell[cell_indexes[a]].remove(self.missing_data_cell[cell_indexes[x]][0])
                    #print('The Updated data in each cell')
                    #for key in self.missing_data_cell:
                    #    print(key + " -> ", self.missing_data_cell[key])
                    
        length = 0
        for a in range(0, len(cell_indexes)):
            length += len(self.missing_data_cell[cell_indexes[a]])
        if length == self.sudoku_size ** 2 :
            self.sector_update_flag[sector_number] = 0
            
    

                
    #iterate through all the entire sector            
    def all_sector_check(self):
        while any(self.sector_update_flag):
            
            for a in range(0, len(self.sector_update_flag)):
                if self.sector_update_flag[a] == 1:
                    self.sector_values_check(a)
                    
        #print('The final cell data')
        #for key in self.missing_data_cell:
        #    print(key + " -> ", self.missing_data_cell[key])
            
    
    def assign_values_back(self):
        for a in range(0,self.sudoku_size):
            for b in range(0,self.sudoku_size):
                for c in range(0,self.sudoku_size):
                    for d in range(0,self.sudoku_size):
                        self.sudoku_matrix[a][b][c][d] = self.missing_data_cell[str(a)+str(b)+str(c)+str(d)]
                        
    def print_values(self):
        print('The final matrix')
        for a in range(0,self.sudoku_size):
            for b in range(0,self.sudoku_size):
                for c in range(0,self.sudoku_size):
                    for d in range(0,self.sudoku_size):
                        print(self.sudoku_matrix[a][c][b][d], end = " ")
                    print(" ", end=" " )
                print("\n")
            print("\n")
        
```
